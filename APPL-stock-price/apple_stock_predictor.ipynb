Lets create a model to predict Apple's stock price using a dataset of Apple's previous stock performance.

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

data = pd.read_csv("AAPL.csv")
print(data.head())
```

Time to visualize the stock data to better determine any trends in Apple's stock price.

```python
figure = go.Figure(data=[go.Candlestick(x=data["Date"],
                                        open=data["Open"], high=data["High"],
                                        low=data["Low"], close=data["Close"])])
figure.update_layout(title = "Apple Stock Price Trend", xaxis_rangeslider_visible=False)
figure.show()
```

Correlation between the features present in the data set.

```python
print(data.corr())
```

Setting up the actual predictor.

```python
from autots import AutoTS
model = AutoTS(forecast_length=5, frequency='infer', ensemble='simple')
model = model.fit(data, date_col='Date', value_col='Close', id_col=None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```
